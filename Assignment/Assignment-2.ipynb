```python
# 12.Employee Salary Based on Experience.
# You are building a system for a Human Resources (HR) department that calculates an
# employee's salary based on their years of experience. The system assigns salary tiers
# based on the number of years an employee has been working. It also offers bonuses for
# employees who have more than 15 years of experience.
# Scenario Breakdown:
# Context 1: Senior Employee
# ● An employee with 10 or more years of experience is classified as a Senior
# Employee. The base salary for such an employee is 80000.
# ● If the employee has more than 15 years of experience they receive a bonus of
# 5000 to their salary.
# Example:
# An employee with 18 years of experience:
# ● They are classified as Senior Employees.
# ● Their base salary is 80000.
# ● Since they have more than 15 years of experience they receive an additional
# 5000 bonus.
# ● The final salary is 85000.
# Context 2: Mid-Level Employee
# ● Employees with 5 to 9 years of experience are classified as Mid-Level
# Employees.
# ● Their base salary is 50000 with no bonus.
# Example:
# An employee with 7 years of experience:
# ● They are classified as a Mid-Level Employee.
# ● Their base salary is 50000.
# ● Since they have fewer than 10 years of experience no bonus is added.
# ● The final salary is 50000.
# Context 3: Junior Employee
# ● Employees with less than 5 years of experience are classified as Junior
# Employees.
# ● Their base salary is 30000 with no bonus.
# Example:
# An employee with 3 years of experience:
# ● They are classified as Junior Employees.
# ● Their base salary is 30000.
# ● No bonus is added.
# ● The final salary is 30000.

experience = int(input("Enter the experience: "))

bonus = 0

# Classification and Salary Logic
if experience >= 10:
    classification = "Senior Employee"
    base_salary = 80000
    if experience > 15:
        bonus = 5000
elif 5 <= experience < 10:
    classification = "Mid-Level Employee"
    base_salary = 50000
else:
    classification = "Junior Employee"
    base_salary = 30000

# Final Calculation
final_salary = base_salary + bonus

# Output
print(f"\n--- HR Salary Calculation ---")
print(f"Classification: {classification}")
print(f"Base Salary: {base_salary}")
if bonus > 0:
    print(f"Bonus: {bonus}")
print(f"Final Salary: {final_salary}")
```

```python
# 15. Tax Calculation for Car Purchase
# Write a program to calculate the tax on a car purchase based on the car brand and its price.
# 1. Mahindra: 5% tax for prices between 7L (7 lakh) and 10L.
# 2. Audi: 10% tax for prices between 10L and 15L.
# 3. Jaguar: 25% tax for prices between 15L and 20L.
# 4. Mercedes: 30% tax for prices between 20L and 25L.
# 5. Input: The car brand and price.
# 6. Output: The calculated tax on the purchase

# Input: Get the car brand and its price
brand = input("Enter the car brand (Mahindra, Audi, Jaguar, Mercedes): ")
price = float(input("Enter the car price (in numbers, e.g., 1200000): "))

# Initialize tax percentage
tax_rate = 0
is_eligible = True

if brand == "Mahindra":
    if 700000 <= price <= 1000000:
        tax_rate = 0.05  # 5%
    else: 
        is_eligible = False

elif brand == "Audi":
    if 1000000 <= price <= 1500000:
        tax_rate = 0.10  # 10%
    else:
        is_eligible = False

elif brand == "Jaguar":
    if 1500000 <= price <= 2000000:
        tax_rate = 0.25  # 25%
    else:
        is_eligible = False

elif brand == "Mercedes":
    if 2000000 <= price <= 2500000:
        tax_rate = 0.30  # 30%
    else:
        is_eligible = False

else:
    print("Error: Unknown or unsupported car brand.")
    is_eligible = False

# Output: Calculate and display the final tax
if is_eligible:
    calculated_tax = price * tax_rate
    print(f"\n--- Tax Invoice Summary ---")
    print(f"Car Brand: {brand}")
    print(f"Car Price: ₹{price:,.2f}")
    print(f"Applied Tax Rate: {tax_rate * 100}%")
    print(f"Calculated Tax Amount: ₹{calculated_tax:,.2f}")
else:
    if brand in ["Mahindra", "Audi", "Jaguar", "Mercedes"]:
        print(f"\nNo tax calculated: The price entered does not fall within {brand}'s designated tax bracket.")
```

```python
# 23.Create Your Own KBC Game
# Design and implement a quiz game inspired by the popular Kaun Banega Crorepati (KBC)
# game show. The aim of this assignment is to test the user's knowledge through a series of
# multiple-choice questions, track their score, and display statistics at the end of the game. The
# game also provides the flexibility to skip any question.
# Instructions:
# 1. Game Structure:
# ○ The game will consist of 5 multiple-choice questions.
# ○ The user will be asked a question with 4 options (A, B, C, D).
# ○ The user can choose to skip any question they do not want to answer.
# 2. Scoring System:
# ○ Points will be awarded for correct answers as follows:
# ■ Question 1 → 1000 points
# ■ Question 2 → 2000 points
# ■ Question 3 → 3000 points
# ■ Question 4 → 5000 points
# ■ Question 5 → 10000 points
# ○ For incorrect answers, no points will be awarded.
# ○ For skipped questions, no points will be awarded, but the game will continue.
# 3. End of Game Statistics:
# ○ At the end of the game, the following statistics will be displayed:
# ■ Total score accumulated from correct answers.
# ■ Number of correct answers provided by the user.
# ■ Number of skipped questions.
# ■ Number of wrong answers
# 4. User Experience:
# ○ At the beginning of the game, ask the user whether they would like to start or
# not the game.
# ○ Provide the option for the user to skip any question at any point..
# 5. Game Ending:
# ○ The game will end when all the questions have been answered or skipped. The
# user should receive their total score and a summary of their performance.


# Stats tracking variables
total_score = 0
correct_answers = 0
wrong_answers = 0
skipped_questions = 0

# Game start prompt
start_game = input("Welcome to KBC! Do you want to start the game? (yes/no): ").strip().lower()

if start_game == "yes":
    print("\n Game Started! You can skip any question by typing 'skip'.")

    # ==========================================
    # QUESTION 1
    # ==========================================
    print("\nQ1: Which is the largest planet in our Solar System?")
    print("  A. Earth")
    print("  B. Jupiter")
    print("  C. Saturn")
    print("  D. Mars")
    
    user_answer = input("Your answer (A/B/C/D or type 'skip'): ").strip().upper()
    
    if user_answer == "SKIP":
        print("Question skipped.")
        skipped_questions = skipped_questions + 1
    elif user_answer == "B":
        print("Correct!")
        total_score = total_score + 1000
        correct_answers = correct_answers + 1
    elif user_answer == "A" or user_answer == "C" or user_answer == "D":
        print("Wrong answer! The correct answer was: B")
        wrong_answers = wrong_answers + 1
    else:
        print("Invalid input. Question marked as skipped.")
        skipped_questions = skipped_questions + 1

    # ==========================================
    # QUESTION 2
    # ==========================================
    print("\nQ2: Who wrote the national anthem of India?")
    print("  A. Bankim Chandra Chatterjee")
    print("  B. Sarojini Naidu")
    print("  C. Rabindranath Tagore")
    print("  D. Mahatma Gandhi")
    
    user_answer = input("Your answer (A/B/C/D or type 'skip'): ").strip().upper()
    
    if user_answer == "SKIP":
        print("Question skipped.")
        skipped_questions = skipped_questions + 1
    elif user_answer == "C":
        print("Correct!")
        total_score = total_score + 2000
        correct_answers = correct_answers + 1
    elif user_answer == "A" or user_answer == "B" or user_answer == "D":
        print("Wrong answer! The correct answer was: C")
        wrong_answers = wrong_answers + 1
    else:
        print("Invalid input. Question marked as skipped.")
        skipped_questions = skipped_questions + 1

    # ==========================================
    # QUESTION 3
    # ==========================================
    print("\nQ3: What is the capital of Australia?")
    print("  A. Sydney")
    print("  B. Melbourne")
    print("  C. Perth")
    print("  D. Canberra")
    
    user_answer = input("Your answer (A/B/C/D or type 'skip'): ").strip().upper()
    
    if user_answer == "SKIP":
        print("Question skipped.")
        skipped_questions = skipped_questions + 1
    elif user_answer == "D":
        print("Correct!")
        total_score = total_score + 3000
        correct_answers = correct_answers + 1
    elif user_answer == "A" or user_answer == "B" or user_answer == "C":
        print("Wrong answer! The correct answer was: D")
        wrong_answers = wrong_answers + 1
    else:
        print("Invalid input. Question marked as skipped.")
        skipped_questions = skipped_questions + 1

    # ==========================================
    # QUESTION 4
    # ==========================================
    print("\nQ4: Which element has the chemical symbol 'O'?")
    print("  A. Gold")
    print("  B. Oxygen")
    print("  C. Iron")
    print("  D. Silver")
    
    user_answer = input("Your answer (A/B/C/D or type 'skip'): ").strip().upper()
    
    if user_answer == "SKIP":
        print("Question skipped.")
        skipped_questions = skipped_questions + 1
    elif user_answer == "B":
        print("Correct!")
        total_score = total_score + 5000
        correct_answers = correct_answers + 1
    elif user_answer == "A" or user_answer == "C" or user_answer == "D":
        print("Wrong answer! The correct answer was: B")
        wrong_answers = wrong_answers + 1
    else:
        print("Invalid input. Question marked as skipped.")
        skipped_questions = skipped_questions + 1

    # ==========================================
    # QUESTION 5
    # ==========================================
    print("\nQ5: In which year did India gain independence?")
    print("  A. 1945")
    print("  B. 1946")
    print("  C. 1947")
    print("  D. 1950")
    
    user_answer = input("Your answer (A/B/C/D or type 'skip'): ").strip().upper()
    
    if user_answer == "SKIP":
        print("Question skipped.")
        skipped_questions = skipped_questions + 1
    elif user_answer == "C":
        print("Correct!")
        total_score = total_score + 10000
        correct_answers = correct_answers + 1
    elif user_answer == "A" or user_answer == "B" or user_answer == "D":
        print("Wrong answer! The correct answer was: C")
        wrong_answers = wrong_answers + 1
    else:
        print("Invalid input. Question marked as skipped.")
        skipped_questions = skipped_questions + 1

    # End of Game Statistics Summary
    print("\n--- Game Over! Here's your performance summary ---")
    print(f"Total Score: {total_score}")
    print(f"Correct Answers: {correct_answers}")
    print(f"Wrong Answers: {wrong_answers}")
    print(f"Skipped Questions: {skipped_questions}")

else:
    print("Thank you! Maybe next time. Goodbye!")
```
